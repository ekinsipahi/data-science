# Stock correllation of nasdaq and AAPL
### This mini project is just calculating the risk aversion of an typical investor that might take
### for a very long time aapl was 1.5x the risk of whole market 
### depending on the 5 year check we did today we can successfully confirm this. with a success rate of %76

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# we should only take the High
nasdaq = pd.read_csv("./NASDAQ.csv")
aapl = pd.read_csv("./AAPL.csv")
aapl.describe()
# nasdaq['Close'] and aapl['Close']
```

```python
X = pd.DataFrame(nasdaq, columns=['Close'])
y = pd.DataFrame(aapl, columns=['Close'])
```

```python
plt.figure(figsize=(10,6))
plt.title("Stock Correllation For risk aversion")
plt.xlabel("Nasdaq")
plt.ylabel("Apple")
plt.xlim(6000, 18200)
plt.ylim(40, 300)
plt.scatter(X, y, alpha=0.3)
```

```python
regression = LinearRegression()
```

```python
regression.fit(X, y)
```

```python
regression.coef_ # Risk is 1.5x of aapl
```

```python
regression.intercept_ # not bad! negative but still raises questions.
```

```python
plt.figure(figsize=(10,6))
plt.title("Stock Correllation For risk aversion")
plt.xlabel("Nasdaq")
plt.ylabel("Apple")
plt.xlim(6000, 18200)
plt.ylim(40, 300)
plt.scatter(X, y, alpha=0.3)
plt.plot(X, regression.predict(X), color='red', linewidth=0.3)
```

```python
regression.score(X, y) # Very good prediction %76 SCORE!
```
